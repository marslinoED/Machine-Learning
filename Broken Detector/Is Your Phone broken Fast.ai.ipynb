```python
from fastbook import *
import cv2
```

```python
path = Path("Phones")
PhoneBroken = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=[Resize(192, method='squish')]
).dataloaders(path, bs=6)

PhoneBroken.show_batch(max_n=6)
```

```python
learn = vision_learner(PhoneBroken, resnet18, metrics=error_rate)
learn.fine_tune(5)
```

```python
is_broken,_,probs = learn.predict(PILImage.create('TEST/BrokenTest.jpg'))
print(f"This is a: {is_broken}.")
print(f"Probability it's a {is_broken}: {max(probs[0], probs[1]):.4f}")


print("______________________")

is_broken,_,probs = learn.predict(PILImage.create('Phones/New/download (0).jpeg'))
print(f"This is a: {is_broken}.")
print(f"Probability it's a {is_broken}: {max(probs[0], probs[1]):.4f}")
```

```python
cap = cv2.VideoCapture(0)

while True:
    success, img = cap.read()
    img = cv2.flip(img,1)
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    
    is_broken,_,probs = learn.predict(img)
    cv2.putText(img, is_broken, (10,40), cv2.FONT_HERSHEY_SIMPLEX,1 , (0, 255, 0), 2)
    
    print(probs[0])
    cv2.imshow("Image", img)
    # Check for the 'Esc' key press
    key = cv2.waitKey(1)
    if key == 27:  # 27 is the keycode for 'Esc'
        break

cap.release()
cv2.destroyAllWindows()
```
