```python
import cv2
import mediapipe as mp
import math
import serial.tools.list_ports
import time
import numpy
import serial
```

```python
ports = serial.tools.list_ports.comports()
serialInst = serial.Serial()


for x in range(len(ports)):
    if(ports[x][1][0:7] == "Arduino"):
        portvar = "COM" + str(ports[0][0][3])
        print("Connecting To Serial Monitor Of: ", portvar)
        
serialInst.baudraten = 9600
serialInst.port = portvar
serialInst.open()
```

```python
def returnTheSortedColors(frame):
    step = int(math.sqrt((frame.shape[0]*frame.shape[1])/1000))
    unique_colors = set()
    height, width, _ = frame.shape
    for y in range(0, height, step):
        for x in range(0, width, step):
            color = tuple(frame[y, x])
            unique_colors.add(color)
    sorted_colors = sorted(unique_colors)
    num = int((len(sorted_colors)/100)*99)
    return sorted_colors[num]


def draw_circle_left_corner(frame, color):
    """
    Draws a circle in the left corner of the frame with the specified color.

    Args:
    - frame: A NumPy array representing an image frame in the format (height, width, channels).
    - color: A tuple representing the color in RGB format.

    Returns:
    - frame_with_circle: A frame with a circle drawn in the left corner.
    """

    # Convert color values to integers
    color = tuple(int(c) for c in color)

    # Draw a circle in the left corner
    radius = 50
    center = (radius, radius)
    cv2.circle(frame, center, radius, color, -1)  # -1 indicates fill the circle

    return frame
k = 0
def Command(stg):
    global k
    k +=1
    print(k)
    
    if(k == 30):
        print("k")
        k = 0
        command = str(stg)
        print(stg)
        serialInst.write(command.encode('utf-8'))
    
def Commanding(stg):

    command = str(stg)
    print(stg)
    serialInst.write(command.encode('utf-8'))
```

```python
# Open video file
video_path = 'Data/Dumbledore vs Voldemort.mp4'
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Unable to open video file.")
    exit()

# Read and process each frame
while True:
    ret, frame = cap.read()

    if not ret:
        break  # Break if there are no more frames
    # Find the lightest color in the frame
    lightest_color_frame = returnTheSortedColors(frame)

    # Draw a circle in the left corner with the lightest color
    frame_with_circle = draw_circle_left_corner(frame, lightest_color_frame)
    Command(str(lightest_color_frame[0])+","+str(lightest_color_frame[1])+","+str(lightest_color_frame[2]))
    # Display the frame with the circle
    cv2.imshow('Frame with Circle', frame_with_circle)

    # Exit if 'q' is pressed
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

# Release video capture object and close windows
cap.release()
cv2.destroyAllWindows()
```
